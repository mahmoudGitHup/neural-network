1.Differentiate $\tanh(x)$ with respect to $x$:
$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

let $u=e^x+e^{-x}$:
$$
\tanh(x)=\frac{e^x+e^{-x}}{u}
$$


using the quotient rule:
    $$
    \frac{d\tanh(x)}{dx}=\frac{(e^x-e^{-x})'u-(e^x-e^{-x})u'}{u^2}
    $$


$$
(e^x-e^{-x})'=e^x+e^{-x}=u
$$
$$
u'=(e^x+e^{-x})=e^x-e^{-x}
$$

then :
    $$
    \frac{d\tanh(x)}{dx}=\frac{u*u-(e^x-e^{-x})(e^x-e^{-x})}{u^2}=\frac{u^2-(e^{2x}-2+e^{-2x})}{u^2}
    $$
$$
u^2=e^{2x}+2+e^{-2x}
$$


then:
    $$
    \frac{u^2-(e^{2x}-2+e^{-2x})}{u^2}=\frac{(e^{2x}+2+e^{-2x})-(e^{2x}-2+e^{-2x})}{e^{2x}+2+e^{-2x}}
    $$
    $$
    (e^{2x}+2+e^{-2x})-(e^{2x}-2+e^{-2x})=4
    $$
as a result:
    $$
    \frac{d\tanh(x)}{dx}=\frac{4}{e^{2x}+2+e^{-2x}}
    $$
